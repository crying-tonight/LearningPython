```python
import matplotlib.pyplot as plt

```

```python
plt.title("Portfolio Diversification")
plt.xlabel("Market Value")
plt.ylabel("Stock Symbol")
Y = [40,30,15,20,25]
X = ["T","VE","APPL","TSLA","KO"]
plt.pie(Y,labels=X)
plt.show()
```

```python
plt.title("Captial Gains")
plt.xlabel("Dollar Amount")
plt.ylabel("Stock Symbol")
Y = [40,30,25,15,35]
X = ["T","VE","TSLA","APPL","KO"]
plt.bar(X,height=Y)
plt.show()

```

```python
plt.title("TSLA")
plt.xlabel("Date")
plt.ylabel("Stock Price")
Y = [450,675,580,675,700]
X = ["1/05","1/10","1/15","1/20","1/25"]
plt.plot(X,Y)
plt.show()
```
